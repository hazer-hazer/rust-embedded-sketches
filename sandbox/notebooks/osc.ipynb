```python
import math
import matplotlib.pyplot as plt

def sawsin(x: float) -> float:
    t = math.fmod(x / (2 * math.pi), 1.0)
    if t > 0.5:
        return -math.sin(x)
    if t <= 0.5:
        return 2.0 * t - 1.0

for i in range(1000):
    x = i / 10
    plt.plot(i, sawsin(x), 'ro')

plt.show()
```
