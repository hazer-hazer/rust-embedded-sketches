```python
import math
import matplotlib.pyplot as plt

def sine(sr: int, freq: float, phase: float, length: int):
    wave_len = sr // freq
    result = [0.0] * length
    phase_offset = wave_len * phase

    for i in range(length):
        result[i] = math.sin((math.pi * 2 * (i + phase_offset)) / wave_len)

    return result

sr = 44100
length = 100
sine1 = sine(sr, 440, 0.0, length)
sine2 = sine(sr, 1000, 0.25, length)
added = [(sine1[i] + sine2[i]) / 2 for i in range(length)]

for x in range(max(len(sine1), len(sine2))):
    plt.plot(x, sine1[x], 'ro')
    plt.plot(x, sine2[x], 'bo')
    plt.plot(x, added[x], 'go')


plt.show()
```
